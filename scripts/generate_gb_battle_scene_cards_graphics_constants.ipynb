```python
from PIL import Image
from numpy import asarray, array_equal, hsplit, vsplit

tiles_image = asarray(Image.open('input/cards_tiles.bmp'))
tiles = [tiles_image[0:8,x:x+8] for x in range(0,tiles_image.shape[1],8)]

cards_image = asarray(Image.open('input/cards.bmp'))
cards_tiles = vsplit(cards_image, cards_image.shape[0] / 8)
for i,cards_tiles_col in enumerate(cards_tiles):
    cards_tiles[i] = hsplit(cards_tiles_col, cards_image.shape[1] / 8)

def findTile(the_tile):
    for i,tile in enumerate(tiles):
        if array_equal(tile, the_tile):
            return i
    return -1

def locGroupToCCode(group):
    ret = "{"
    for location in group:
        ret += f"{{{location[0]}, {location[1]}, {findTile(cards_tiles[location[1]][location[0]])}}},"
    ret += "}"
    return ret

def locGroupsToCCode(groups):
    ret = "{"
    for group in groups:
        ret += locGroupToCCode(group) + ","
    ret += "}"
    return ret
```
